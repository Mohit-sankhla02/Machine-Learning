# Making Changes to the Data

```python
df.head(5)
```

**Now suppose we want to create a new attribute total that stores the sum of all the numerical stats**

```python
df["total"] = df.HP + df.Attack + df.Defense + df["Sp. Atk"] + df["Sp. Def"] + df.Speed
```

```python
df.head(5)
```

```python
# we can also write as

df["Total"] = df.iloc[:, 4:10].sum(axis=1)
df.head(5)
```

### Dropping a column

```python
df = df.drop(columns=['total'])
df.head(5)
```

```python
# putting total in the left of HP
cols = list(df.columns)
df = df[cols[0:4] + [cols[-1]] + cols[4:12]]
df.head(5)
```

```python
df.to_csv('modified_pokemon_data.csv', index=False)
```

## Filtering Data

```python
df.loc[(df['Type 1'] == 'Grass') & (df['Type 2'] == 'Poison')]
# this will show us the method 

# this will print out the entries with type 1 as grass and type 2 as only Poison
```

```python
# we can do if either one of them is useful
df[(df['Type 1'] == "Grass") | (df['Type 2'] == 'Poison')]
```

```python
# now if we want to use that filtered data for something
new_df = df[(df['Type 1'] == "Grass") & (df['Type 2'] == "Poison") & (df['Total'] > 400)]

new_df

```

## ADVANCED TEXTUAL FILTERING

**USING REGULAR EXPRESSIONS**

```python
import re

df.loc[df['Type 1'].str.contains('fire|grass', flags=re.I, regex=True)] # all names starting with pi
```

```python
df.loc[df['Name'].str.contains('^pi[a-z]*', flags=re.I, regex=True)]
```

## CONDITIONAL CHANGES

```python
df.loc[df['Type 1'] == "Fire", 'Type 1'] = 'Flamer'
df

```

```python
# reverting the changes back
df.loc[df['Type 1'] == "Flamer", "Type 1"] = "Fire"
df
```

```python
df.loc[df['Total'] > 500, ['Generation', 'Legendary']] = ['Test 1', 'Test 2']

df
```

## Aggregate Statistics (groupby)

```python
df = pd.read_csv('modified_pokemon_data.csv')
df.groupby(['Type 1']).mean()
```

#### Now we can see that we get the average of every value on the basis of type1

#### we can now get some useful information that we need 

```python
df.groupby(['Type 1']).mean().sort_values('Defense', ascending=False)
```

#### We can see that on the basis of defense, we can get the highest level of defense

```python
df.groupby(['Type 1']).mean().sort_values('Attack', ascending=False)
```

#### We can see that the dragon has the highest attack of them all

### Now what if we needed to find the number of pokemons in each category

```python
df['count'] = 1
df
```

```python
# now with this count column, we can find out the number of pokemon of each type by using sum property
new_count = df.groupby(['Type 1', 'Type 2']).sum()['count']
new_count

new_count.to_csv('useful_data.csv')
```

# Working with large amounts of data

```python
new_df = pd.DataFrame(columns=df.columns)
for df in pd.read_csv('modified_pokemon_data.csv', chunksize=5):
    result = df.groupby(['Type 1']).count()
    
    new_df = pd.concat([new_df, result])
    
    
new_df
```
